# 2.3 Data Cleaning and Normalization

```python
from sklearn.preprocessing import MinMaxScaler

## Drop duplicate rows and rows with missing values
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)

## Separate label column before normalizing
label_col = "Label"  # adjust if your label column has a different name
attack_col = "Attack"  # adjust if your label column has a different name
X = df.drop(columns=[label_col, attack_col])
y = df[label_col]
```

```python
print(X.columns)
```

```python
## Split features by VarType
categorical_features = features_df[features_df["VarType"] == "Categorical"]["Feature"].tolist()
quantitative_features = features_df[features_df["VarType"] == "Quantitative"]["Feature"].tolist()

# Filter to only columns that exist in X (exclude label, etc.)
categorical_features = [f for f in categorical_features if f in X.columns]
quantitative_features = [f for f in quantitative_features if f in X.columns]

X_cat = X[categorical_features]
X_quant = X[quantitative_features]
```

```python
features_df_var = pd.read_csv("./data/NetFlow_v3_Features_VarType.csv")
print(features_df_var)
```

## Treating Qualitative/Categorical Features

```python
# Drop IPv4 addresses
X_cat = X_cat.drop(columns=["IPV4_SRC_ADDR", "IPV4_DST_ADDR"])
print(X_cat.columns)
```

```python
## Binary Decomposition (before encoding, flags are already integers)
flag_cols = ["TCP_FLAGS", "CLIENT_TCP_FLAGS", "SERVER_TCP_FLAGS"]
for col in flag_cols:
    for bit, name in enumerate(["FIN","SYN","RST","PSH","ACK","URG"]):
        C = X_cat[col].astype(int).apply(lambda x: x >> bit)
        X_cat[f"{col}_{name}"] = C & 1
X_cat = X_cat.drop(columns=flag_cols)
```

```python
# Label Encode
from sklearn.preprocessing import LabelEncoder
import joblib

label_encoders = {}
for col in X_cat.columns:
    le = LabelEncoder()
    X_cat[col] = le.fit_transform(X_cat[col].astype(str))
    label_encoders[col] = le

joblib.dump(label_encoders, "./train_params/label_encoders.pkl")
```

```python
# One-hot encode low cardinality columns
low_card_cols = [col for col in X_cat.columns if X_cat[col].nunique() < 20]
X_cat = pd.get_dummies(X_cat, columns=low_card_cols)
```

```python
# Frequency Encode port numbers
for col in ["L4_SRC_PORT", "L4_DST_PORT"]:
    # value_count of each column divided by sum of value_counts from each column
    freq = X_cat[col].value_counts(normalize=True)
    X_cat[col] = X_cat[col].map(freq)
```

```python
print(X_cat.head())
```

## Treating Quantiative Features

```python
## Summary Statistics; showing extreme skewness on quantitative features.
## Transforming is necessary to mitigate skewness.
print(X_quant.describe())
print("\nSkewness:")
print(X_quant.skew())
```

```python
import numpy as np
from sklearn.preprocessing import RobustScaler
import joblib

X_quant = X_quant.replace([np.inf, -np.inf], np.nan)
X_quant = X_quant.dropna(axis=1, how="all")  # drop cols that are entirely NaN
X_quant = X_quant.fillna(0)                  # fill remaining NaNs

# Log-Transform to safely handle 0 (log1p: log(1+x))
X_quant = X_quant.apply(lambda col: np.log1p(col) if (col >= 0).all() else col)

# Clip outliers
upper = X_quant.quantile(0.99)
X_quant = X_quant.clip(upper=upper, axis=1)

# Scaler
scaler = RobustScaler()
X_quant = pd.DataFrame(scaler.fit_transform(X_quant), columns=X_quant.columns)

# save scaler for later use.
joblib.dump(scaler, "./train_params/robust_scaler.pkl")
print(X_quant)

```

```python
## Processed Data
X_processed = pd.concat([X_cat.reset_index(drop=True), X_quant.reset_index(drop=True)], axis=1)
print(X_processed.head())
```

```python
# Uncomment below lines to save as csv
X_processed.to_csv("./data/X_processed.csv", index=False)
y.to_csv("./data/y.csv", index=False)
```

```python
import pandas as pd
# Use this if necessary
X_processed = pd.read_csv("./data/X_processed.csv")
y = pd.read_csv("./data/y.csv")
```

```python
# View proportion of Benign (0) and Malignant (1) results
print(y.value_counts())
print(y.value_counts(normalize=True).mul(100).round(2).astype(str) + "%")
```

```python
from sklearn.model_selection import train_test_split

## Generate Train, Test
seed = 42

# First split off test set (20%)
X_train, X_test, y_train, y_test = train_test_split(
    X_processed, y, test_size=0.2, random_state=42, stratify=y
)
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=50, max_depth=20, class_weight={0:1, 1:10}, random_state=42)
rf.fit(X_train, y_train)

# Predict
y_proba = rf.predict_proba(X_test)[:, 1]
threshold = 0.5
y_pred = (y_proba >= threshold).astype(int)
```

```python
## Classification Report
from sklearn.metrics import classification_report, roc_auc_score, precision_recall_curve, confusion_matrix
import matplotlib.pyplot as plt

print(classification_report(y_test, y_pred))
print("AUC:", roc_auc_score(y_test, y_proba))

print(confusion_matrix(y_test, y_pred))

precision, recall, thresholds = precision_recall_curve(y_test, y_proba)

plt.plot(thresholds, precision[:-1], label='Precision')
plt.plot(thresholds, recall[:-1], label='Recall')
plt.xlabel('Threshold')
plt.legend()
plt.show()
```

## Creating Test Datasets from simulated attacks on Wireshark

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, RobustScaler
import glob

## Load CSVs
brute_hydra_ftp_v2_df = pd.read_csv("./output-csv/brute-hydra-ftp-v2.csv")
brute_hydra_ssh_df = pd.read_csv("./output-csv/brute-hydra-ssh.csv")
dos_syn_flood_df = pd.read_csv("./output-csv/dos-syn-flood.csv")
recon_nmap_all_ports_df = pd.read_csv("./output-csv/recon-nmap-all-ports.csv")
recon_nmap_sCsV_df = pd.read_csv("./output-csv/recon-nmap-sCsV.csv")

## Load feature type metadata
features_df = pd.read_csv("./data/NetFlow_v3_Features_VarType.csv")
categorical_features = features_df[features_df["VarType"] == "Categorical"]["Feature"].tolist()
quantitative_features = features_df[features_df["VarType"] == "Quantitative"]["Feature"].tolist()

categorical_cols = [f for f in categorical_features if f in X.columns]
quantitative_cols = [f for f in quantitative_features if f in X.columns]
```

```python
import joblib

scaler = joblib.load("./train_params/robust_scaler.pkl")  # load the training scaler
label_encoders = joblib.load("./train_params/label_encoders.pkl") # load the training label encoders


def clean_pcap_data(df, attack_label=""):
    x = df.copy()
    
    ## Drop duplicates
    x.drop_duplicates(inplace=True)
    x.dropna(inplace=True)

    ## Split data into Categorical vs quantitative
    x_cat = x[categorical_features]
    x_quant = x[quantitative_features]

    ## Process categorical data
    x_cat = x_cat.drop(columns=["IPV4_SRC_ADDR", "IPV4_DST_ADDR"], errors="ignore")
    flag_cols = ["TCP_FLAGS", "CLIENT_TCP_FLAGS", "SERVER_TCP_FLAGS"]
    for col in [c for c in flag_cols if c in x_cat.columns]:
        for bit, name in enumerate(["FIN","SYN","RST","PSH","ACK","URG"]):
            x_cat[f"{col}_{name}"] = x_cat[col].astype(int).apply(lambda x: x >> bit) & 1
    x_cat = x_cat.drop(columns=[c for c in flag_cols if c in x_cat.columns])

    label_encoders = {}
    for col in x_cat.columns:
        le = LabelEncoder()
        x_cat[col] = le.fit_transform(x_cat[col].astype(str))
        label_encoders[col] = le

    low_card_cols = [col for col in x_cat.columns if x_cat[col].nunique() < 20]
    x_cat = pd.get_dummies(x_cat, columns=low_card_cols)

    for col in ["L4_SRC_PORT", "L4_DST_PORT"]:
        if col in x_cat.columns:
            freq = x_cat[col].value_counts()
            x_cat[col] = x_cat[col].map(freq)

    # 6. Quantitative processing (mirrors notebook exactly)
    x_quant = x_quant.replace([np.inf, -np.inf], np.nan)
    x_quant = x_quant.dropna(axis=1, how="all")
    x_quant = x_quant.fillna(0)
    x_quant = x_quant.apply(lambda col: np.log1p(col) if (col >= 0).all() else col)
    upper = x_quant.quantile(0.99)
    x_quant = x_quant.clip(upper=upper, axis=1)
    x_quant = pd.DataFrame(scaler.transform(x_quant), columns=x_quant.columns)  # transform with training scaler

    # 7. Combine
    x_processed = pd.concat([x_cat.reset_index(drop=True), x_quant.reset_index(drop=True)], axis=1)
    y_test = pd.DataFrame(1, index=range(x_processed.shape[0]), columns=["Label"]) # all ones to determine malicious
    z_test = pd.DataFrame(attack_label, index=range(x_processed.shape[0]), columns=["Label"])
    return x_processed, y_test, z_test

x_recon_nmap, y_recon_nmap, z_recon_nmap = clean_pcap_data(recon_nmap_all_ports_df, "Reconnaissance")
x_recon_sCsV, y_recon_sCsV, z_recon_sCsV = clean_pcap_data(recon_nmap_sCsV_df, "Reconnaissance")
x_dos_synflood, y_dos_synflood, z_dos_synflood = clean_pcap_data(dos_syn_flood_df, "DoS")
x_brute_hydraftp, y_brute_hydraftp, z_brute_hydraftp = clean_pcap_data(brute_hydra_ftp_v2_df, "Generic")
x_brute_hydrassh, y_brute_hydrassh, z_brute_hydrassh = clean_pcap_data(brute_hydra_ssh_df, "Generic")
```
